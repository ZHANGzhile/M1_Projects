```python
import sys
sys.path.append('../Class')
sys.path.append('../utils')


%matplotlib inline
%load_ext autoreload

import matplotlib.pyplot as plt
import numpy as np
from transformations import TanH, Sigmoid
```

```python
# Create instances of TanH and Sigmoid
tanh = TanH()
sigmoid = Sigmoid()

# Define a range of values
x = np.linspace(-10, 10, 100)

# Calculate TanH and Sigmoid values
tanh_values = tanh.forward(x)
sigmoid_values = sigmoid.forward(x)

# Plot TanH values
plt.figure(figsize=(10, 5))
plt.plot(x, tanh_values, label='TanH')
plt.plot(x, sigmoid_values, label='Sigmoid')
plt.title('TanH and Sigmoid functions')
plt.xlabel('x')
plt.ylabel('Function value')
plt.legend()
plt.grid(True)
plt.show()
```
